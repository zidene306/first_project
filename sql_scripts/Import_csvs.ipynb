```python
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
import getpass  # To get the password without showing the input
password = getpass.getpass("Please, input your SQL password: ")

bd = "project1_amazon"
connection_string = 'mysql+pymysql://root:' + password + '@localhost/'+bd
engine = create_engine(connection_string)

# -------------------------
# DB CONNECTION
# -------------------------
# engine = create_engine("mysql+pymysql://root:password@localhost/ecommerce_db")

# -------------------------
# LOAD CSV FILES
# -------------------------
products = pd.read_csv("products.csv")
users = pd.read_csv("users.csv")
categories = pd.read_csv("categories.csv")
product_categories = pd.read_csv("product_categories.csv")
reviews = pd.read_csv("reviews.csv")
product_images = pd.read_csv("product_images.csv")

# -------------------------
# LOAD ORDER (IMPORTANT FOR FOREIGN KEYS)
# -------------------------

# 1. Parent tables first
products.to_sql("products", engine, if_exists="append", index=False)
users.to_sql("users", engine, if_exists="append", index=False)
categories.to_sql("categories", engine, if_exists="append", index=False)

# 2. Junction tables
product_categories.to_sql("product_categories", engine, if_exists="append", index=False)

# 3. Dependent tables
reviews.to_sql("reviews", engine, if_exists="append", index=False)
product_images.to_sql("product_images", engine, if_exists="append", index=False)

# -------------------------
# DONE
# -------------------------
print("✅ All CSVs loaded into MySQL successfully!")
```
